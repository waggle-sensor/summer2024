## Locating Individual Patches in an Image

````python
# Getting the coordinates for the white pixels in the image
# Second array represents x-axis, First array represents y-axis

white_scan = np.nonzero(scan1)
white_scan
````

````python
# Min max coordinates for x and y coordinates

x_min = white_scan[1].min()
x_max = white_scan[1].max()
y_min = white_scan[0].min()
y_max = white_scan[0].max()

x_min, x_max, y_min, y_max
````

````python
# Creating a tuple for the coordinates

list_coords = tuple(zip(white_scan[1], white_scan[0]))

# Filtering out redundant coordinates using the list function
list_coords = list(dict.fromkeys(list_coords))

print(list_coords)
````

````python
# Performing a DBSCAN clustering algorithm to sort out individual patches

from sklearn.cluster import DBSCAN

points = np.array(list_coords)

# hard coded values, could definitely be optimized
distance, min_samples = 51, 1

db = DBSCAN(eps = distance, min_samples = min_samples).fit(points)

# Printing out number of clusters
print("Num Clusters:", db.labels_.max() + 1)
````

````python
coords = []
dist = []

for i in range(db.labels_.max() + 1):
#     coord = np.round(np.mean(points[db.labels_ == i], axis = 0)).astype(int)  # coordinates for generating same sized patches

    # distance spread between x,y coordinates for individual patches (small numbers indicate smaller patches, larger numbers indicate larger patches)
    dist_spread = np.round(np.max(points[db.labels_ == i], axis = 0) - np.min(points[db.labels_ == i], axis = 0)).astype(int)

    # coordinates for generating different sized patches
    coord = np.round(np.min(points[db.labels_ == i], axis = 0) + dist_spread / 2).astype(int)

    coords.append((coord[0].item(), coord[1].item()))
    dist.append((dist_spread[0].item(), dist_spread[1].item()))

print("Coords:")
print(coords)
print("\nDist:")
print(dist)
````

````python
# small baseline offset for better visualization purposes so boxing a patch does not hide a patch
offset = 15

patch_size = []

for i in range(db.labels_.max() + 1):
    width = dist[i][0] * 1.5 + offset
    height = dist[i][1] * 1.5 + offset

    # square patches
    length = max(width, height)
    patch_size.append(length)

    # rectangle patches
#     patch_size.append((width, height)

print("Square Sizes:")
print(patch_size)
````

````python
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots(figsize=(100, 100), dpi=60)

# Display the image
ax.imshow(scan1)

# Create patches

for i in range(len(patch_size)):
    length = patch_size[i]

#   Square patches
    rect_patch = patches.Rectangle((coords[i][0] - length / 2, coords[i][1] - length / 2),
                                   length, length, linewidth = 7, edgecolor = 'r', facecolor = 'none')

#     Rectangle patches
#     rect_patch = patches.Rectangle((coords[i][0] - width / 2, coords[i][1] - height / 2),
#                                    width, height, linewidth = 7, edgecolor = 'r', facecolor = 'none')
    ax.add_patch(rect_patch)

plt.show()
````
