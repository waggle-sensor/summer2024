## Create "sage_query_-1h_sense.csv"

```python
df_test = sage_data_client.query(start='-1h', tail=1)
```

```python
print(sorted(df_test['name'].dropna().unique()))
```

```python
names = ['env.pressure', 'env.relative_humidity', 'env.temperature']
df_new = df_test[df_test['name'].isin(names)].dropna(axis=1, how='all')
df_new.sample(frac=1.0).to_csv('sage_query_-1h_sense.csv')
```
